```python
emotions_1=pd.read_csv("dataset/Categorical/Ses01F_script01_1_e4_cat.txt", sep=":", names = ["file_name", "emotion"])
emotions_2=pd.read_csv("dataset/Categorical/Ses01F_script01_2_e4_cat.txt", sep=":", names = ["file_name", "emotion"])
emotions_3=pd.read_csv("dataset/Categorical/Ses01F_script01_3_e4_cat.txt", sep=":", names = ["file_name", "emotion"])
emotions = pd.concat([emotions_1, emotions_2, emotions_3], axis = 0)
emotions
```

```python
emotions.loc[:,"emotion"] = emotions.emotion.replace(to_replace="; \(\)", value="", regex=True)
emotions.loc[:,"emotion"] = emotions.emotion.replace(to_replace="; ", value="", regex=True)
emotions.loc[:,"file_name"] = emotions.file_name.replace(to_replace=" ", value="", regex=True)

emotions.emotion.unique()
```

```python
emotion_groupby = emotions.groupby(by="emotion")
emotions[~emotions.loc[:,'emotion'].str.contains("Other")].groupby(by="emotion").count()
```

```python
emotions_fil = emotions[~emotions.loc[:,'emotion'].str.contains("Other")].groupby(by="emotion").head(7).reset_index(drop=True)
emotions_fil
```

```python
hand_crafted = []
sigs_diff_length = []
sigs = []
time_steps = []
names = []

for file in emotions_fil[emotions_fil.file_name.str.contains("Ses01F_script01_1")].file_name:
    (sig,rate) = core.load("dataset/wav/Ses01F_script01_1/{}.wav".format(file), sr = 8000)
    sigs_diff_length.append(sig)
    time_steps.append(len(sig))
    names.append("{}".format(file))

for file in emotions_fil[emotions_fil.file_name.str.contains("Ses01F_script01_2")].file_name:
    (sig,rate) = core.load("dataset/wav/Ses01F_script01_2/{}.wav".format(file), sr = 8000)
    sigs_diff_length.append(sig)
    time_steps.append(len(sig))
    names.append("{}".format(file))

for file in emotions_fil[emotions_fil.file_name.str.contains("Ses01F_script01_3")].file_name:
    (sig,rate) = core.load("dataset/wav/Ses01F_script01_3/{}.wav".format(file), sr = 8000)
    sigs_diff_length.append(sig)
    time_steps.append(len(sig))
    names.append("{}".format(file))

time_step = max(time_steps)

for sig in sigs_diff_length:
    sig = np.pad(sig, (0, time_step - len(sig)), mode = "constant")
    sig = np.reshape(sig, (-1,1)) 
    sigs.append(sig)
print(len(names), len(sigs), sigs[10].shape)
```

```python
sigs
```

```python
sigs_df = pd.DataFrame([names, sigs])
```

```python
sigs_df = sigs_df.T
```

```python
sigs_df.rename(columns={0:"file_name", 1:"sigs"}, inplace=True)
sigs_df
```

```python
x_y = pd.merge(sigs_df, emotions_fil, how="inner", left_on = 'file_name', right_on = 'file_name' ).set_index(keys="file_name" )
```
